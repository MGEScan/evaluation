```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
mpi2 =  (1037,462,409)
mpi4 = (684,328,263)
mpi8 = (610,281,278)
mpi12 = (318,151,156)

ind = np.arange(N)  # the x locations for the groups
width = 0.25       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar([ 0.25,  1.50,  2.75], mpi2, width, color='r')
rects2 = ax.bar([ 0.5,  1.75,  3.0], mpi4, width, color='y')
rects3 = ax.bar([ 0.75,  2.0,  3.25], mpi8, width, color='b')
rects4 = ax.bar([1, 2.25, 3.5], mpi12, width, color='g')

# add some text for labels, title and axes ticks
ax.set_ylabel('Elapsed Time (sec)')
ax.set_title('Elapsed Time by Number of MPI Processes for nonLTR')
ax.set_xticks([0.75, 2.0, 3.25])
ax.set_xticklabels(('d.melanogaster','c.elegans','a.thaliana'))

ax.legend((rects1[0], rects2[0], rects3[0], rects4[0]), ('MPI2', 'MPI4', 'MPI8', 'MPI12'))


def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.show()
```
